```python
mask_array = np.asarray(mask_machine)
pprint(mask_array)
pprint(mask_array[0,0])
```

```python
def RGBinequality(RGB):
    # print(RGB)
    r,g,b = RGB
    # print(r,g,b)
    return (r!=g) or (g!=b)

RGBinequality(mask_array[0,0])
```

```python
false_indices = [(i, j)
                for i, row in enumerate(mask_array)
                for j, pixel in enumerate(row)
                if RGBinequality(pixel)]
false_indices
```

```python
def get_RGBinequal_indices(mask_machine):
    mask_array = np.asarray(mask_machine)
    false_indices = [(i, j)
                    for i, row in enumerate(mask_array)
                    for j, pixel in enumerate(row)
                    if RGBinequality(pixel)]
    return false_indices
```

# Check all masks

```python
for _,_,mask_machine,_,_,filename in taskent_dataset:
    false_indices = get_RGBinequal_indices(mask_machine)

    print(filename, len(false_indices))
    if false_indices:
        pprint(false_indices)
```
